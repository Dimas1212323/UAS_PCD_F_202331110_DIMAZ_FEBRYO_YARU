```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Baca gambar

```python
img = cv2.imread('foto_diri_filtering.jpg')
```

### Median filtering (fungsi bawaan OpenCV)

```python
median = cv2.medianBlur(img, 5)
```

### Mean filtering (manual pakai kernel rata-rata)

```python
kernel = np.ones((3, 3), np.float32) / 9
mean = cv2.filter2D(img, -1, kernel)
```

### Tampilkan hasil

```python
titles = ['Gambar Asli', 'Median Filter', 'Mean Filter (Manual)']
images = [img, median, mean]
```

```python
plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
